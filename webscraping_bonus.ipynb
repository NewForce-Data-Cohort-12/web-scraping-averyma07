1. Navigate to https://www.billboard.com/charts/hot-100/. Using BeautifulSoup, extract out the This Week, artist, song, Last Week, Peak Position, and Weeks on Chart values into a pandas DataFrame. Hint: The HTML for the number one ranked song is slightly different from that of the rest of the songs.

```python
response = requests.get('https://www.billboard.com/charts/hot-100/', headers=headers)
```

```python
response.status_code
```

```python
soup = bs(response.text)
```

```python
this_week = [x.get_text().strip() for x in soup.find_all('span', class_='c-label a-font-basic u-font-size-22@desktop u-font-size-19@mobile-max u-line-height-33px')]
```

```python
this_week.insert(0, '1')
```

```python
title_tags = soup.find_all('h3', class_='c-title a-font-basic u-letter-spacing-0010 u-max-width-397 lrv-u-font-size-16 lrv-u-font-size-14@mobile-max u-line-height-22px u-word-spacing-0063 u-line-height-normal@mobile-max a-truncate-ellipsis-2line lrv-u-margin-b-025 lrv-u-margin-b-00@mobile-max')
```

```python
song_titles = [x.get_text().strip() for x in title_tags]
```

```python
artist_tags = soup.find_all('span', class_='c-label a-no-trucate a-font-secondary u-font-size-15 u-font-size-13@mobile-max u-line-height-18px@mobile-max u-letter-spacing-0010 u-line-height-21px a-children-link-color-black a-children-link-color-brand-secondary:hover lrv-a-children-link-decoration-underline:hover lrv-u-display-block a-truncate-ellipsis-2line u-max-width-397 u-max-width-230@tablet-only u-max-width-300@mobile-max')
```

```python
artists = [x.get_text().strip() for x in artist_tags]
```

```python
stat_tags = soup.find_all('span', class_='c-label u-font-family-secondary@mobile-max u-font-family-basic@tablet u-font-weight-800@tablet u-font-size-12 u-line-height-normal lrv-u-padding-tb-00@mobile-max u-min-width-30px u-width-auto@mobile-max u-min-width-auto@mobile-max lrv-u-margin-r-050@mobile-max')
```

```python
stats = [x.get_text().strip() for x in stat_tags]
```

```python
last_week = []
peak = []
weeks_on_chart = []
```

```python
for i, stat in enumerate(stats):
    if i % 6 == 0:
        last_week.append(stat)
    elif i % 6 == 1:
        peak.append(stat)
    elif i % 6 == 2:
        weeks_on_chart.append(stat)
```

```python
hot_100 = pd.DataFrame({'this_week':this_week, 'song':song_titles, 'artist':artists, 'last_week':last_week, 'peak_position':peak, 'weeks_on_chart': weeks_on_chart})
```
